###  **1. O que é paradigma procedural?**
    
  - <p>O nome do paradigma se faz pois o mesmo se baseia em procedimentos que precisam ser executados em uma determinada sequência pré determinada pelo usuário ou por um sistema estabelecido e acrescentando procedimentos que organizam o código em blocos reutilizáveis

#### **1.1 O que são procedimentos?**
  - <p> Além disso, os procedimentos podem se chamar mutuamente. Isso significa que um procedimento pode invocar outro para realizar uma subtarefa, tornando o código mais organizado e modular. Essa característica é fundamental para a construção de programas complexos e bem estruturados.

####  **1.2 Resumindo paradigma procedural**
  - <p> A programação procedural é derivada de uma outra programação chamada de imperativa, o principal foco dessa programação denominada procedural é que o mesmo pode ser dividido em blocos para serem reutilizados futuramente inclusive por outros procedimentos.

  - Procedimentos  e modularidade
	Modularidade: utiliza canais de entrada e de saída estritamente definidos, usualmente acompanhados de regras claras sobre quais tipos de entrada e saída.

###**2. Resumo introdutório**
<p> A linguagem C, uma das mais influentes e utilizadas da história da computação, foi desenvolvida para ser simples e eficiente, proporcionando aos seus utilizadores controle diretamente sobre os recursos da parte física(hardware). C tem foco em manipulação de memória, controle de fluxo e estrutura de dados, sendo amplamente empregada ao uso de desenvolvimento de sistemas operacionais, aplicações de baixo nível e compiladores. Apesar de sua estrutura simples, oferece a capacidade para suportar uma ampla gama de aplicações, tornando- se assim a base de muitas outras linguagens modernas.
<p>O fato de ser uma linguagem focada na questão da programação procedural, seu código é dividido em blocos, facilitando o reuso e a dinâmica de como o algoritmo interage com si mesmo, esse foi um ponto muito forte pelo qual a linguagem C permanece sendo usada, além do controle do hardware como já foi dito, fazendo com que algoritmos bem feito em C proporcionem uma ótima qualidade para um projeto, sua história percorre por muitos pontos, a seguir será apresentado toda o percurso dessa linguagem de programação.


####**3.2. História Geral**

**3.2.1 - Características Gerais do nascimento da linguagem**
<p> Criada por Dennis Ritchie em 1972, um cientista da computação americano, na empresa AT&T Bell Labs (Laboratórios Bell) uma das intituições mais prestigiadas no desenvolvimento de tecnologias de telecomunicação e computação, como uma evolução da linguagem B, que se derivava de BCPL(Basic Combined Programming Language). Tinha como objetivo fornecer uma linguagem de programação eficiente para o desenvolvimento do sistema operacional Unix.
<p>A necessidade de ambientes portáteis e eficientes para implementar sistemas operacionais, que pudesse rodar em diferentes máquinas foi a principal motivação para criação de C. Ao contrário de outras linguagens da mesma época, que eram muito específicas para determinadas máquinas, ela foi projetada para ser maleável e permitir o desenvolvimento de programas eficientes em termos de uso de memória e desempenho.
Tradicionalmente, não possui logotipo oficial amplamente reconhecido, ao contrário de linguagens contemporâneas que com frequência adotam visuais mais fortes e marcantes. No entanto, o símbolo mais comum associado à linguagem é uma letra "C" em tipografia simples.


**3.2.2 - Criadores da linguagem**
<p> O principal desenvolvedor foi Dennis Ritchie, que trabalhou no desenvolvimento do Unix ao lado de Ken Thompson, que desenvolveu a linguagem B, que antecedeu diretamente o C. Ambos trabalhavam nos laboratórios Bell, e a necessidade já mencionanda antes, levou à criação do C. Além deles, tiveram outros contribuintes para o aprimoramento da linguagem ao longo dos anos, como Alan Snyder, Steven C. Johnson e Michael Lesk.
<p>Com isso, em 1978 Dennis Ritchie e Brian Kernighan lançaram um livro intitulado “The C programming Language”, o mesmo servia como uma documentação da linguagem e auxiliava usuários e entusiastas da linguagem C com tutoriais e como aplicar funções e manipular variáveis por exemplo, o livro foi chama chamado de “K&R C” e foi considerada como uma parte básica da linguagem, futuras edições iriam abranger mais tópicos da linguagem como os tipos de dados “struct”, “long int” e “unsigned int”.


**3.2.3 - Versões do C (C K&R - C23)**
<p> A linguagem C ao longo dos anos passou por várias versões, contando com revisões e padronizações, cada uma delas buscando melhorias e novas funcionalidades.
<p> Abaixo, uma breve descrição do caminho que foi feito através de cada versão dessa linguagem:

  **→ C K&R:** Essa foi a primeira versão formal da linguagem C popularizada pelo livro *The C Programming Language*, de Brian Kernighan e Dennis Ritchie, publicado no ano de 1978. Popularmente conhecida como C K&R(Kernighan e Ritchie), o livro serviu como a primeira especificação informal da linguagem, trouxe para serem introduzidos conceitos básicos como a biblioteca padrão de entrada/saída (`stdio.h`), dados como `struct`, `long int`, e operadores de atribuição como `+=` e `-=`. Tornou-se amplamente aceito, e por muitos anos foi referência, sendo utilizada até mesmo após a introdução de padrões subsequentes, como o ANSI C, descrito a seguir.
  
  **→ ANSI C e ISO C:** No ano de 1983, o intituto norte-americano de padrões (ANSI) montou o comitê X3J11 para padronizar a linguagem C. A culminância desse trabalho se deu na publicação do padrão ANSI C em 1989, que formalmente é conhecido como ANSI X3.159-1989, chamado também de C89. Versão que foi depois adotada pela ISO (Organização Internacional de Padronização), sendo referida como C90. Seu objetivo central era produzir um subconjunto do K&R C, implementando funcionalidades não oficiais adicionadas por desenvolvedores. O ANSI C/ISO C trouxe esse padrão e maior portabilidade de programas em C, proporcionando assim a fácil compatibilidade entre diferentes plataformas e compiladores.
  
  **→ C99:** Após a padronização ANSI e apresentado em 1999, trouxe melhorias incluindo suporte para variáveis de comrimento dinâmico, funções matemáticas, funções em linhas e novos tipos de dados, como o long, long int e complex, também foi adicionado suporte a comentários de uma linha com //(Puxado do C++).
  
  **→ C11:** Lançada em 2011, seu foco era melhorar a segurança e a ambiguidade, como novos recursos como dados atômicos e um novo conjunto de bibliotecas.

  **→ C17:** Lançada em 2018, trouxe apenas pequenas revisões e correções, sem adicionar novos recursos realmente significativos para a linguagem.

  **→ C23:** Essa revisão ainda esta em desenvolvimento, e promete trazer inovações, mantendo a compatibilidade com versões anteriores refinando a linguagem para novas necessidades.


####**3.3. Como o C trabalha com a Linguagem Procedural e Porque é usada nesse paradigma**  
<p> Ser uma linguagem de programação procedural significa que os programas são compostos de procedimentos ou funções. Nesse tipo de paradigma, o foco está em dividir o programa em módulos ou blocos lógicos, chamados funções, que executam tarefas, ou procedimentos que podem ser reutilizados. Isso ajuda a organizar o código de forma mais compacta e eficiente.
<p>

###**4. Características gerais da Linguagem C**

####**4.1 Simplicidade e eficiência**
<p> Traço marcante e notável de C, é a simplicidade. Possui um conjunto pequeno, comparado a outra linguagens, de palavras-chave e operadores, o que possibilita a facilidade no apredizado e na compreensão. Contudo, o fato de ser simples não compromete a eficiência já que C permite poder direto sobre os recursos de hardware, como memória e processamento, se tornando assim ideal para programar sistemas.

####**4.2 Portabilidade**
<p> A portabilidade de C é uma das razões do sucesso da linguagem. Programas escritos nessa linguagem podem ser executados e compilados por várias plataformas sem a necessidade de grandes alterações no código-fonte. Isso se dá pelo fato de C ter sido desenvolvido com abstrações que o difere dos detalhes específicos do harware, mas sem perder a eficiência.

####**4.3 Controle direto de memória**
<p> Oferece controle direto sobre a memória através do uso de ponteiros, isso é um importante diferencial tendo em vista as outras linguagens de alto nível. Ponteiros permitem a manipulação do endereço de memória e facilitam a implementação de estrutura de dados mais complexas, como grafos, listas encadeadas e árvores. Esse tipo de controle é muito útil no desenvolvimento de sistemas operacionais e drivers.

####**4.4 Extensibilidade**
<p> Mesmo planejado para ser uma linguagem básica e eficiente, oferece meios que permitem a criação de bibliotecas e funções personalizadas. Facilitando assim a adaptação da linguagem às necessidades específicas de resolução de problemas, permitindo que os desenvolvedores criem extensões de acordo com suas necessidades, mas mantendo a base da linguagem intacta.

####**4.5 Suporte a biblioteca padrão**
<p>A biblioteca padrão é chamada de biblioteca de rotina, pois contém as operações comuns e mais utilizadas nos programas em geral, sendo assim operações básicas para um programa funcionar corretamente.
<p>Cabeçalho é o local onde estão as características das funções, porém cada compilador pode diferenciar com base no ANSI C.
<p>Algumas funções padrões da biblioteca C como o gets() e scanf() às vezes geram defeitos como o buffer overflow e o strtok(), um analisador léxico primitivo, porém se torna complicado de se manipular e por muitas vezes geram erros.
<p>A linguagem C na sua gênese não possuía funcionalidades nativas, nem a de entrada e nem a de saída, com isso gerando dúvidas e muitos usuários criaram em conjunto ideias e trabalhos para ter a melhorar, com isso foi criado a biblioteca < stdio.h> (standard input-output), uma biblioteca que possuem as funções mais básicas de entrada e saída de dados.
<p>Consiste em 24 cabeçalhos no padrão ANSI C, cada cabeçalho consiste em ferramentas padrões e básicas que auxiliam a linguagem na construção de blocos de códigos ou algoritmos, alguns cabeçalhos do ANSI C são: < ctype.h>, < locale.h>, < stdio.h>, < string.h> e < float.h>.

####**4.6 Estabilidade e padrões**
<p> Mesmo criada há mais de quatro décads, permaneceu relativamente estável, com poucas mudanças realmente significativas ao longo das várias revisões. Isso garante que os programas que foram escritos em outras versões mais antigas da linguagem ainda podem ser executadas nas atuais versões, preservando assim o código ao longo do tempo.
<p>Fazendo assim a linguagem C atravessar gerações e permanecer sendo utilizada em tecnologias atuais, como por exemplo máquinas hospitalares, sua ampla liberdade para manipular a memória é um ponto positivo dependendo do projeto estabelecido, portanto essa linguagem de programação ainda vai continuar em evidência em algumas áreas da tecnologia da informação.

```python
# prompt: Crie um algoritmo na linguagem C que respeite o paradigma procedural e simples, códigos em blocos que possuam reusabilidade

#include <stdio.h>

// Função para calcular a soma de dois números
int somar(int a, int b) {
  return a + b;
}

// Função para calcular a diferença entre dois números
int subtrair(int a, int b) {
  return a - b;
}

// Função para calcular o produto de dois números
int multiplicar(int a, int b) {
  return a * b;
}

// Função para calcular a divisão de dois números
float dividir(int a, int b) {
  if (b == 0) {
    printf("Erro: Divisão por zero!\n");
    return 0;
  }
  return (float)a / b;
}

int main() {
  int num1, num2, escolha;

  printf("Digite o primeiro número: ");
  scanf("%d", &num1);

  printf("Digite o segundo número: ");
  scanf("%d", &num2);

  printf("\nEscolha uma operação:\n");
  printf("1 - Somar\n");
  printf("2 - Subtrair\n");
  printf("3 - Multiplicar\n");
  printf("4 - Dividir\n");
  printf("Digite sua opção: ");
  scanf("%d", &escolha);

  switch (escolha) {
    case 1:
      printf("Soma: %d\n", somar(num1, num2));
      break;
    case 2:
      printf("Subtração: %d\n", subtrair(num1, num2));
      break;
    case 3:
      printf("Multiplicação: %d\n", multiplicar(num1, num2));
      break;
    case 4:
      printf("Divisão: %.2f\n", dividir(num1, num2));
      break;
    default:
      printf("Opção inválida!\n");
  }

  return 0;
}
```
